```python
import numpy as np
import matplotlib.pyplot as plt

# x 값 정의
x = np.linspace(-5, 5, 400)

# 함수 정의 리스트
functions = [
    (3*x + 1, "y = 3x + 1"),
    (x**2, "y = x^2"),
    (np.exp(x), "y = e^x"),
    (np.maximum(0, x), "ReLU"),
    (1 / (1 + np.exp(-x)), "Sigmoid"),
    (np.tanh(x), "tanh")
]

# Figure 생성
fig = plt.figure(figsize=(8, 6))

# add_subplot()으로 2행 3열 배치
for i, (y, label) in enumerate(functions, 1):
    ax = fig.add_subplot(2, 3, i)   # (행, 열, 위치)
    ax.plot(x, y, label=label)
    ax.set_title(label)
    ax.grid(True)
    ax.legend()
    if label == "y = e^x":
        ax.set_ylim(-1, 50)  # 지수함수 범위 제한

fig.suptitle("다양한 그래프", fontsize=16)
plt.tight_layout()
plt.show()
```

### y축 공유

```python
import numpy as np
import matplotlib.pyplot as plt

# x 값 정의
x = np.linspace(-5, 5, 400)

# 함수 정의 리스트
functions = [
    (3*x + 1, "y = 3x + 1"),
    (x**2, "y = x^2"),
    (np.exp(x), "y = e^x"),
    (np.maximum(0, x), "ReLU"),
    (1 / (1 + np.exp(-x)), "Sigmoid"),
    (np.tanh(x), "tanh")
]

# Figure 생성
fig = plt.figure(figsize=(10, 6))

axes = []  # 각 subplot 저장
for i, (y, label) in enumerate(functions, 1):
    ax = fig.add_subplot(2, 3, i)
    ax.plot(x, y, label=label)
    ax.set_title(label)
    ax.grid(True)
    ax.legend()
    axes.append((ax, y, label))

# === 행별 y축 범위 맞추기 ===
# 첫 번째 행 (subplot 1,2,3)
row1_data = np.hstack([y for _, y, _ in axes[:3]])
row1_min, row1_max = row1_data.min(), row1_data.max()
# 지수함수는 값이 너무 크므로 보기 좋게 제한
row1_max = min(row1_max, 50)
for ax, _, _ in axes[:3]:
    ax.set_ylim(row1_min, row1_max)

# 두 번째 행 (subplot 4,5,6)
row2_data = np.hstack([y for _, y, _ in axes[3:]])
row2_min, row2_max = row2_data.min(), row2_data.max()
for ax, _, _ in axes[3:]:
    ax.set_ylim(row2_min, row2_max)

# 전체 제목
fig.suptitle("다양한 그래프 (행별 y축 공유)", fontsize=16)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 준비
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
data = np.random.randn(1000)

# Figure 생성
fig = plt.figure(figsize=(6, 4))
fig.suptitle("fig.add_subplot()을 이용한 2행 3열 그래프 예제", fontsize=16)

# (1,1): Sine Wave
ax1 = fig.add_subplot(2, 3, 1)
ax1.plot(x, y, color='b')
ax1.set_title("Sine Wave")

# (1,2): Cosine Wave
ax2 = fig.add_subplot(2, 3, 2)
ax2.plot(x, z, color='r')
ax2.set_title("Cosine Wave")

# (1,3): Scatter Plot
ax3 = fig.add_subplot(2, 3, 3)
ax3.scatter(x, y, color='g')
ax3.set_title("Scatter Plot")

# (2,1): Histogram
ax4 = fig.add_subplot(2, 3, 4)
ax4.hist(data, bins=30, color='purple')
ax4.set_title("Histogram")

# (2,2): Box Plot
ax5 = fig.add_subplot(2, 3, 5)
ax5.boxplot(data)
ax5.set_title("Box Plot")

# (2,3): Bar Chart
ax6 = fig.add_subplot(2, 3, 6)
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]
ax6.bar(categories, values, color='orange')
ax6.set_title("Bar Chart")

# 레이아웃 조정
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.01, 0.99, 400)
y1 = np.log(x)
y2 = np.log(1 - x)

fig = plt.figure(figsize=(7,3))

# log(x)
ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(x, y1, label="log(x)")
ax1.set_ylim(-5, 0.5)
ax1.set_title("log(x)")
ax1.grid(True)
ax1.legend()

# log(1-x)
ax2 = fig.add_subplot(1, 2, 2, sharey=ax1)
ax2.plot(x, y2, label="log(1-x)")
ax2.set_title("log(1-x)")
ax2.grid(True)
ax2.legend()

plt.tight_layout()
plt.show()
```

### fig.add_subplot()에서 y축 공유

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.01, 0.99, 400)
y1 = np.log(x)
y2 = np.log(1 - x)

fig = plt.figure(figsize=(7,3))

ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(x, y1, label="log(x)")
ax1.set_title("log(x)")
ax1.grid(True)
ax1.legend()

# 두 번째 subplot, y축 공유
ax2 = fig.add_subplot(1, 2, 2, sharey=ax1)
ax2.plot(x, y2, label="log(1-x)")
ax2.tick_params(labelleft=False) # y축 눈금 제거
ax2.set_title("log(1-x)")
ax2.grid(True)
ax2.legend()

plt.tight_layout()
plt.show()
```
