## 함수 plt.subplot()

```python
import matplotlib.pyplot as plt

plt.subplot(2,2,1)
plt.plot([1,2,3],[1,4,9])
plt.title("제곱 그래프", fontsize=10)   # 각 subplot 제목
plt.subplot(2,2,4)
plt.plot([1,2,3],[1,2,3])
plt.title("직선 그래프", fontsize=10)

plt.suptitle("Subplot 그래프", fontsize=16, fontweight="bold")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 준비
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
data = np.random.randn(1000)

plt.figure(figsize=(6, 4))
# 전체 제목
plt.suptitle("plt.subplot()으로 그린 2행 3열 그래프 예제", fontsize=16)

# (1,1): 선 그래프
plt.subplot(2, 3, 1)  # 2행 3열 중 첫 번째
plt.plot(x, y, 'b-')
plt.title("Sine Wave")

# (1,2): 코사인 곡선
plt.subplot(2, 3, 2)
plt.plot(x, z, 'r-')
plt.title("Cosine Wave")

# (1,3): 산점도
plt.subplot(2, 3, 3)
plt.scatter(x, y, c='g', s=.5)
plt.title("Scatter Plot")

# (2,1): 히스토그램
plt.subplot(2, 3, 4)
plt.hist(data, bins=30, color='purple')
plt.title("Histogram")

# (2,2): 박스플롯
plt.subplot(2, 3, 5)
plt.boxplot(data)
plt.title("Box Plot")

# (2,3): 막대그래프
plt.subplot(2, 3, 6)
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]
plt.bar(categories, values, color='orange')
plt.title("Bar Chart")

# 레이아웃 조정
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# x 범위 (0,1) 사이
x = np.linspace(0.01, 0.99, 400)
y1 = np.log(x)
y2 = np.log(1 - x)

plt.figure(figsize=(6,3))

# log(x)
plt.subplot(1, 2, 1)
plt.plot(x, y1, label="log(x)")
plt.ylim(-5, 0.5)   # y축 동일하게
plt.title("log(x)")
plt.grid(True)
plt.legend()

# log(1-x)
plt.subplot(1, 2, 2)
plt.plot(x, y2, label="log(1-x)")
plt.ylim(-5, 0.5)   # y축 동일하게
plt.title("log(1-x)")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

### plt.subplot()에서 y축 공유

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.01, 0.99, 400)
y1 = np.log(x)
y2 = np.log(1 - x)

plt.figure(figsize=(6,3))

# 첫 번째 subplot
ax1 = plt.subplot(1, 2, 1)
ax1.plot(x, y1, label="log(x)")
ax1.set_title("log(x)")
ax1.grid(True)
ax1.legend()

# 두 번째 subplot, y축 공유
ax2 = plt.subplot(1, 2, 2, sharey=ax1)
ax2.plot(x, y2, label="log(1-x)")
ax2.tick_params(labelleft=False)
ax2.set_title("log(1-x)")
ax2.grid(True)
ax2.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# x 값 정의
x = np.linspace(-5, 5, 400)

# 함수 정의
y1 = 3*x + 1                    # y = 3x + 1
y2 = x**2                       # y = x^2
y3 = np.exp(x)                  # y = e^x
y4 = np.maximum(0, x)           # ReLU
y5 = 1 / (1 + np.exp(-x))       # Sigmoid
y6 = np.tanh(x)                 # tanh

# subplot: 2행 3열
fig = plt.figure(figsize=(6, 5))

# 1
plt.subplot(2, 3, 1)
plt.plot(x, y1, label="y = 3x + 1")
plt.legend()
plt.grid(True)
# 2
plt.subplot(2, 3, 2)
plt.plot(x, y2, label="y = x^2")
plt.legend()
plt.grid(True)
# 3
plt.subplot(2, 3, 3)
plt.plot(x, y3, label="y = e^x")
plt.ylim(-1, 50)   # 보기 좋게 제한
plt.legend()
plt.grid(True)

# 4
plt.subplot(2, 3, 4)
plt.plot(x, y4, label="ReLU")
plt.legend()
plt.grid(True)
# 5
plt.subplot(2, 3, 5)
plt.plot(x, y5, label="Sigmoid")
plt.legend()
plt.grid(True)
# 6
plt.subplot(2, 3, 6)
plt.plot(x, y6, label="tanh")
plt.legend()
plt.grid(True)

fig.suptitle("다양한 그래프", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# x 값 정의
x = np.linspace(-5, 5, 400)

# 함수 정의
y1 = 3*x + 1                    # y = 3x + 1
y2 = x**2                       # y = x^2
y3 = np.exp(x)                  # y = e^x
y4 = np.maximum(0, x)           # ReLU
y5 = 1 / (1 + np.exp(-x))       # Sigmoid
y6 = np.tanh(x)                 # tanh

# Figure
plt.figure(figsize=(6, 5))

# 첫 번째 행
plt.subplot(2, 3, 1)
plt.plot(x, y1, label="y = 3x + 1"); plt.legend(); plt.grid(True)
plt.ylim(-5, 50)
plt.subplot(2, 3, 2)
plt.plot(x, y2, label="y = x^2"); plt.legend(); plt.grid(True)
plt.ylim(-5, 50)
plt.subplot(2, 3, 3)
plt.plot(x, y3, label="y = e^x"); plt.legend(); plt.grid(True)
plt.ylim(-5, 50)

# 두 번째 행
plt.subplot(2, 3, 4)
plt.plot(x, y4, label="ReLU"); plt.legend(); plt.grid(True)
# 두 번째 행 y축 공유 (강제)
plt.ylim(-1.5, 5)
plt.subplot(2, 3, 5)
plt.plot(x, y5, label="Sigmoid"); plt.legend(); plt.grid(True)
plt.ylim(-1.5, 5)
plt.subplot(2, 3, 6)
plt.plot(x, y6, label="tanh"); plt.legend(); plt.grid(True)
plt.ylim(-1.5, 5)

# 전체 제목
plt.suptitle("다양한 그래프 (plt.subplot 스타일, 행별 y축 공유)", fontsize=16)
plt.tight_layout()
plt.show()
```
