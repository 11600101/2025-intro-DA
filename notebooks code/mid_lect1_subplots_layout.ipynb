## 함수 plt.subplots()

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(2,2)

ax[0,0].plot([1,2,3],[1,2,3])
ax[0,0].set_title("선 그래프", fontsize=8)   # 각 subplot 제목
ax[0,1].bar([1,2,3],[3,2,1])
ax[0,1].set_title("막대 그래프", fontsize=8)   # 각 subplot 제목
ax[1,0].scatter([1,2,3],[3,2,1])
ax[1,0].set_title("산점도 그래프", fontsize=8)   # 각 subplot 제목
ax[1][1].hist([1,2,2,2,3,3], bins=[.5, 1.5, 2.5, 3.5], rwidth=.95)
ax[1,1].set_title("돗수분포 그래프", fontsize=8)   # 각 subplot 제목
fig.suptitle("Subplot 그래프", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2,2)
for i, axi in enumerate(ax.flat):
    axi.plot([1,2,3],[n*(i+1) for n in [1,2,3]])
    axi.set_title(f"Plot {i+1}")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
x = np.linspace(0,10,100)
y1, y2, y3 = np.sin(x), np.cos(x), np.tan(x)

fig, ax = plt.subplots(2,2, figsize=(6,5))
ax[0,0].plot(x,y1); ax[0,0].set_title("Sine")
ax[0,1].plot(x,y2); ax[0,1].set_title("Cosine")
ax[1,0].plot(x,y3); ax[1,0].set_title("Tangent")
ax[1,1].hist(y1, bins=20); ax[1,1].set_title("Histogram")
fig.suptitle("삼각함수 그래프")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 준비
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)

# 2행 3열 subplot 생성
fig, axes = plt.subplots(2, 3, figsize=(7, 4))

# 전체 제목 추가
fig.suptitle("다양한 그래프 예제 (2행 3열 Subplots)", fontsize=16)

# (0,0): 선 그래프
axes[0,0].plot(x, y, color='b')
axes[0,0].set_title("Sine Wave")

# (0,1): 코사인 곡선
axes[0,1].plot(x, z, color='r')
axes[0,1].set_title("Cosine Wave")

# (0,2): 산점도
axes[0,2].scatter(x, y, color='g')
axes[0,2].set_title("Scatter Plot")

# (1,0): 히스토그램
data = np.random.randn(1000)
axes[1,0].hist(data, bins=30, color='purple')
axes[1,0].set_title("Histogram")

# (1,1): 박스플롯
axes[1,1].boxplot(data)
axes[1,1].set_title("Box Plot")

# (1,2): 막대그래프
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]
axes[1,2].bar(categories, values, color='orange')
axes[1,2].set_title("Bar Chart")

# plt.tight_layout(rect=[0, 0, 1, 0.95])  # 전체 제목 공간 확보
plt.tight_layout()  # 전체 제목 공간 확보
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 준비
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)

# 2행 3열 subplot 생성
fig, axes = plt.subplots(2, 3, figsize=(7, 4))

# 전체 제목 추가
fig.suptitle("다양한 그래프 예제 (2행 3열 Subplots)", fontsize=16)

# (0,0): 선 그래프
axes[0,0].plot(x, y, color='b')
axes[0,0].set_title("Sine Wave")

# (0,1): 코사인 곡선
axes[0,1].plot(x, z, color='r')
axes[0,1].set_title("Cosine Wave")

# (0,2): 산점도
axes[0,2].scatter(x, y, color='g')
axes[0,2].set_title("Scatter Plot")

# (1,0): 히스토그램
data = np.random.randn(1000)
axes[1,0].hist(data, bins=30, color='purple')
axes[1,0].set_title("Histogram")

# (1,1): 박스플롯
axes[1,1].boxplot(data)
axes[1,1].set_title("Box Plot")

# (1,2): 막대그래프
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]
axes[1,2].bar(categories, values, color='orange')
axes[1,2].set_title("Bar Chart")

# 공통 X, Y 라벨 추가
fig.supxlabel("공통 X축 라벨", fontsize=12)
fig.supylabel("공통 Y축 라벨", fontsize=12)

# 제목과 라벨 겹치지 않도록 조정
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

### plt.subplots()에서 y축 공유

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.01, 0.99, 400)
y1 = np.log(x)
y2 = np.log(1 - x)

fig, axes = plt.subplots(1, 2, figsize=(7,3), sharey=True)

axes[0].plot(x, y1, label="log(x)")
axes[0].set_title("log(x)")
axes[0].grid(True)
axes[0].legend()

axes[1].plot(x, y2, label="log(1-x)")
axes[1].set_title("log(1-x)")
axes[1].grid(True)
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# x 값 정의
x = np.linspace(-5, 5, 400)

# 함수 정의 리스트
functions = [
    (3*x + 1, "y = 3x + 1"),
    (x**2, "y = x^2"),
    (np.exp(x), "y = e^x"),
    (np.maximum(0, x), "ReLU"),
    (1 / (1 + np.exp(-x)), "Sigmoid"),
    (np.tanh(x), "tanh")
]

# 2행 3열 subplots 생성
fig, axes = plt.subplots(2, 3, figsize=(7, 6))

# flatten으로 1차원 배열로 변환
axes = axes.flatten()

# 각 subplot에 함수 그리기
for ax, (y, label) in zip(axes, functions):
    ax.plot(x, y, label=label)
    ax.set_title(label)
    ax.grid(True)
    ax.legend()
    if label == "y = e^x":
        ax.set_ylim(-1, 50)  # 지수함수는 제한 필요

fig.suptitle("다양한 그래프", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# x 값 정의
x = np.linspace(-5, 5, 400)

# 함수 정의 리스트
functions = [
    (3*x + 1, "y = 3x + 1"),
    (x**2, "y = x^2"),
    (np.exp(x), "y = e^x"),
    (np.maximum(0, x), "ReLU"),
    (1 / (1 + np.exp(-x)), "Sigmoid"),
    (np.tanh(x), "tanh")
]

# 2행 3열 subplot, 행별로 y축 공유
fig, axes = plt.subplots(2, 3, figsize=(7, 6), sharey='row')

# flatten()으로 1차원 배열로 변환 → 반복문에 맞게
axes = axes.flatten()

# 그래프 그리기
for ax, (y, label) in zip(axes, functions):
    ax.plot(x, y, label=label)
    ax.set_title(label)
    ax.grid(True)
    ax.legend()
    # if label == "y = e^x":
    #     ax.set_ylim(-1, 50)  # 지수함수는 제한 필요

# 전체 제목
fig.suptitle("다양한 그래프 (행별 y축 공유)", fontsize=14)
# 레이아웃 조정 (제목 공간 확보)
plt.tight_layout()
plt.show()
```
