```python
plt.subplot(2, 2, 1)
plt.plot([1, 2, 3], [4, 4.5, 6])
plt.title("직선")

plt.subplot(2, 2, 2)
plt.plot([1, 2, 3], [1, 4, 9])
plt.title("제곱")

plt.subplot(2, 2, 3)
plt.scatter([1, 2, 3], [5, 4, 3])
plt.title("산점도")

plt.subplot(2, 2, 4)
plt.bar([1, 2, 3], [6, 1, 5])
plt.title("막대 그래프")

plt.suptitle("subplot 그래프", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
plt.subplot(2, 2, 1)
plt.plot([1,2,3],[1,2,3])
plt.title("선 그래프", fontsize=8)   # 각 subplot 제목

plt.subplot(2, 2, 2)
plt.bar([1,2,3],[3,2,1])
plt.title("막대 그래프", fontsize=8)   # 각 subplot 제목

plt.subplot(2, 2, 3)
plt.scatter([1,2,3],[3,2,1])
plt.title("산점도 그래프", fontsize=8)   # 각 subplot 제목

plt.subplot(2, 2, 4)
plt.hist([1,2,2,2,3,3], bins=[.5, 1.5, 2.5, 3.5], rwidth=.90)
plt.title("돗수분포 그래프", fontsize=8)   # 각 subplot 제목

plt.suptitle("전체 그래프", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2, 2)

x = np.arange(5)

ax[0, 0].plot(x, x)
ax[0, 0].set_title("y = x")

ax[0, 1].plot(x, 2*x)
ax[0, 1].set_title("y = 2x")

ax[1, 0].plot(x, x**2)
ax[1, 0].set_title("y = x**2")

ax[1, 1].plot(x, x**3)
ax[1, 1].set_title("y = x**3")

fig.suptitle("subplot 그래프", fontsize=16)
#plt.suptitle("subplot 그래프", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2, 2)

x = np.arange(5)

ax[0][0].plot(x, x)
ax[0][0].set_title("y = x")

ax[0][1].plot(x, 2*x)
ax[0][1].set_title("y = 2x")

ax[1, 0].plot(x, x**2)
ax[1, 0].set_title("y = x**2")

ax[1, 1].plot(x, x**3)
ax[1, 1].set_title("y = x**3")

fig.suptitle("subplot 그래프", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2,2)

ax[0,0].plot([1,2,3],[1,2,3])
ax[0,0].set_title("선 그래프", fontsize=8)   # 각 subplot 제목

ax[0,1].bar([1,2,3],[3,2,1])
ax[0,1].set_title("막대 그래프", fontsize=8)   # 각 subplot 제목

ax[1,0].scatter([1,2,3],[3,2,1])
ax[1,0].set_title("산점도 그래프", fontsize=8)   # 각 subplot 제목

ax[1][1].hist([1,2,2,2,3,3], bins=[.5, 1.5, 2.5, 3.5], rwidth=.95)
ax[1,1].set_title("돗수분포 그래프", fontsize=8)   # 각 subplot 제목

fig.suptitle("전체 그래프", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2, 2)

for i, axi in enumerate(ax.flat):
  axi.plot([1, 2, 3], [n*(i+1) for n in [1, 2,3]])
  axi.set_title(f'Plot {i+1}')

plt.tight_layout()
plt.show()
```
